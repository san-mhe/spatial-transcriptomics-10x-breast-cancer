```python
#leiden_r0_5 will be used in this notebook.

#Spatial neighborhood enrichment

from pathlib import Path
import scanpy as sc
import numpy as np
import squidpy as sq


adata = sc.read_h5ad("../data/processed/visium_breast_cancer_blockA_section1_clustered.h5ad")
adata.obs["leiden_r0_5"] = adata.obs["leiden_r0_5"].astype("category")



sq.gr.spatial_neighbors(adata, coord_type="grid")


sq.gr.nhood_enrichment(
    adata,
    cluster_key="leiden_r0_5"
)


sq.pl.nhood_enrichment(
    adata,
    cluster_key="leiden_r0_5",
    figsize=(6, 6)
)
```

```python
#Spatial co-occurrence

sq.gr.co_occurrence(
    adata,
    cluster_key="leiden_r0_5"
)

sq.pl.co_occurrence(
    adata,
    cluster_key="leiden_r0_5",
    clusters=["0", "1", "5"]  
)
```

```python
Spatial co-occurrence analysis revealed that cluster 1 forms a highly self-enriched domain with limited intermixing, consistent with a spatially coherent epithelial/tumour region. Cluster 5 also exhibited strong short-range self-enrichment, suggestive of a structured immune or stromal niche. 
In contrast, cluster 0 displayed weaker but broader co-occurrence patterns, indicating a more diffuse or interface-like domain.
```
